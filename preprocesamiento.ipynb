```python
#borrar columnas inecesarias
df = df.drop(['Year','Day of Year','Month'], axis=1)
```

```python
#rellenar datos vacios con el fin de post-procesar posteriormente
df = df.fillna(0)
```

```python
df.head()
```

```python
#cambiar nombre de variable
data = df['magnitud']
df = df.assign(sismo=data)
```

```python
#ubicar vabriable categorica para la binarizaci√≥n para la tecnica de Arboles de Decisi√≥n
df.loc[df['sismo'] >0,'sismo']=1
```

```python
df.head()
```

```python
#reordenar columnas
df = df.reindex(columns=['Day','Hour of day','Latitude','Longitude','magnitud','Profundidad','Tec','Ne/cm-3','O+','N+','hmF2','foF2','VTEC','sismo'])
```

```python
#Modificar parametros para realizar graficas
df['magnitud'] = df['magnitud'].map(float)
```

```python
#condiciones de busqueda de datos del dataset
df2 = df[(df['Latitude']==16.84492)&(df['magnitud']>0)]
```

```python
#librerias necesarias para la elaboraci√≥n de graaficas
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

```python
#guardar variables para la elaboracio√≥n de gr√°ficas
x = df2['hmF2']
y = df2['VTEC']
z = df2['magnitud']
```

```python
#creaci√≥n de gr√°ficas en cuestion del analisis de los dias 1-8 de septiembre del 2021 sismo de Acapulco se puede ajustar a cualquier localidad cambiando la latitud
fig=plt.figure(figsize=(8,6))
axes = plt.axes(projection="3d")
axes.scatter3D(x,y,z,color="red")
axes.set_title("Acapulco 2021 3D earthquake analysis",fontsize=14,fontweight="bold")
axes.set_xlabel("hmF2")
axes.set_ylabel("Total electron content")
axes.set_zlabel("Magnitud")
plt.tight_layout()
plt.savefig('aca.png')
plt.show()
```
