```python
#librerias necesarias para la elaboraci칩n de graaficas
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
#guardar variables para la elaboracio칩n de gr치ficas

x = df['foF2']
y = df['hmF2']
z = df['Tec']
```

```python
fig=plt.figure(figsize=(10,10))
axes = plt.axes(projection="3d")
axes.scatter3D(x,y,z,color="blue")
axes.set_title("ANALISIS 3D DEL 1 DE MAYO AL 5 DE MAYO DEL  2023 NOTO JAP칍N",fontsize=14,fontweight="bold")
axes.set_xlabel("foF2")
axes.set_ylabel("hmF2")
axes.set_zlabel("Tec")
plt.tight_layout()
plt.savefig('3D.png')
plt.show()
```

```python
plt.figure(figsize=(12,8))
X1 = df['Hour of day']
plt.plot(X1,x,'--r')
plt.title("Grafica TOTAL ELECTRON CONTENT")
plt.xlabel('DOY')
plt.ylabel('TEC')
plt.savefig('TOTAL ELECTRON.png')
plt.show()
```
